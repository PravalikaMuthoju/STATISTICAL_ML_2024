Read the data with pandas and find features and target variables

```python
import pandas as pd

# Load the data
df = pd.read_csv('/content/archive (9).zip')

# Check the first few rows to understand the data
print(df.head())

# Define features and target
X = df[['YearsExperience']]  # Feature (Years of Experience)
y = df['Salary']             # Target (Salary)
```

```python
import matplotlib.pyplot as plt

# Plot the data
plt.scatter(X, y, color='red', label='Data points')
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predicting for the entire dataset to plot the best fit line
y_pred = model.predict(X)

# Plot the best fit line
plt.scatter(X, y, color='red', label='Data points')
plt.plot(X, y_pred, color='blue', label='Best fit line')
plt.title('Best Fit Line (Linear Regression)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Predictions for the test set
y_test_pred = model.predict(X_test)

# Calculate MSE and MAE
mse_20 = mean_squared_error(y_test, y_test_pred)
mae_20 = mean_absolute_error(y_test, y_test_pred)

print(f'MSE for 20% test size: {mse_20}')
print(f'MAE for 20% test size: {mae_20}')
```

```python
# Split the data (75% train, 25% test)
X_train_25, X_test_25, y_train_25, y_test_25 = train_test_split(X, y, test_size=0.25, random_state=42)

# Create and train the model for 25% split
model_25 = LinearRegression()
model_25.fit(X_train_25, y_train_25)

# Predictions for the 25% test set
y_test_pred_25 = model_25.predict(X_test_25)

# Calculate MSE and MAE
mse_25 = mean_squared_error(y_test_25, y_test_pred_25)
mae_25 = mean_absolute_error(y_test_25, y_test_pred_25)

print(f'MSE for 25% test size: {mse_25}')
print(f'MAE for 25% test size: {mae_25}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

data_dict = {
    'Feature1': [1.2, 2.4, 3.5, 4.1, 1.9, 3.7, 2.1, 3.3, 2.8, 4.0],
    'Feature2': [2.3, 1.9, 2.7, 3.0, 2.8, 2.5, 3.2, 2.4, 3.1, 3.3],
    'Feature3': [3.1, 2.8, 1.5, 3.6, 2.5, 1.9, 2.2, 3.0, 1.8, 2.7],
    'Feature4': [4.2, 3.5, 2.9, 4.8, 3.2, 4.0, 4.1, 4.5, 3.6, 4.9],
    'Target': [15.6, 13.1, 12.4, 18.2, 14.1, 16.0, 14.5, 17.3, 14.9, 19.2]
}

df = pd.DataFrame(data_dict)

X = df[['Feature1', 'Feature2', 'Feature3', 'Feature4']]
y = df['Target']

plt.figure(figsize=(12, 8))
for i, feature in enumerate(X.columns):
    plt.subplot(2, 2, i+1)
    plt.scatter(df[feature], y, label=f'{feature} vs Target')
    plt.xlabel(feature)
    plt.ylabel('Target')
    plt.grid(True)
    plt.legend()
plt.tight_layout()
plt.show()

def evaluate_model(test_size):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    return mse, mae

```
